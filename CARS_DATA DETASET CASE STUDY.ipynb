# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(8,6))
data['Buyer Gender'].value_counts().plot(kind='bar',color=['blue','orange'])
plt.title('MALE VS FEMALE BUYERS',fontsize=15)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid()
```

```python
#From the above-plotted graph, we can conclude that the number of female buyers who participated in the sales is greater than male buyers.
```
